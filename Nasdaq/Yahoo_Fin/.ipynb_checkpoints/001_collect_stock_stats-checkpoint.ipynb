```python
import pandas as pd
import json
import yahoo_fin.stock_info as yf
import numpy as np
import time
import threading
```

```python
nasdaq_df = pd.read_csv('nasdaq_screener.csv')
```

### Convert Last Price column into float

```python
nasdaq_df["Last Sale"] = nasdaq_df["Last Sale"].apply(lambda x : float(x[1:])) # removing $ sign
```

### Read available (on wealth simple) stocks data

```python
available_stocks = pd.read_json("curated_stocks_all.json")
```

```python
available_stocks = available_stocks.rename(columns={'symbol' : 'Symbol', 'name' : 'Name'})
```

```python
wealthsimple_stocks = nasdaq_df[nasdaq_df['Symbol'].isin(available_stocks['Symbol'])]
```

### keep only symbol, price, name, volume columns

```python
columns_to_keep = ['Symbol', 'Name', 'Last Sale','Volume']
wealthsimple_stocks = wealthsimple_stocks.loc[:, columns_to_keep]
wealthsimple_stocks = wealthsimple_stocks.rename(columns={'Last Sale' : 'Price'})
```

### Add new columns (P/E, P/B, P/S, EV/EBITDA, EV/REV, RV Score) and percentile

```python
rv_columns = [
    "P/E", "PE Percentile",
    "P/B", "PB Percentile",
    "P/S", "PS Percentile",
    'EV/EBITDA', 'EV/EBITDA Percentile',
    'EV/Revenue', 'EV/Revenue Percentile',
    'RV Score'
]

# for column in rv_columns:
#     wealthsimple_stocks[column] = [np.NaN] * len(wealthsimple_stocks_try)
```

```python
# Reset index and drop index column
wealthsimple_stocks = wealthsimple_stocks.reset_index().drop(columns='index')
```

### call yahoo_fin api stats on each symbol and populate data
### get_stats_valuation returns dataframe, so make first column its index and then retrieve data

```python
# for i in range(len(wealthsimple_stocks)):

lock = threading.Lock()

stock_stats = stock_stats.set_index(stock_stats.columns[0])
starting = 3726

count = 0

for i in range(starting, len(wealthsimple_stocks)):
    symbol = wealthsimple_stocks.loc[i, 'Symbol']
    # acquire lock wait for function to complete
    lock.acquire()
    stock_stats = yf.get_stats_valuation(symbol)
    lock.release() # release the lock
    stock_stats = stock_stats.set_index(stock_stats.columns[0])
    try:
        wealthsimple_stocks.loc[i, 'P/E'] = float(stock_stats.loc['Trailing P/E', 1])
    except:
        wealthsimple_stocks.loc[i, 'P/E'] = np.NaN
    
    try:
        wealthsimple_stocks.loc[i, 'P/S'] = float(stock_stats.loc['Price/Sales (ttm)', 1])
    except:
        wealthsimple_stocks.loc[i, 'P/S'] = np.NaN
        
    try:
        wealthsimple_stocks.loc[i, 'P/B'] = float(stock_stats.loc['Price/Book (mrq)', 1])
    except:
        wealthsimple_stocks.loc[i, 'P/B'] = np.NaN
    
    try:
        wealthsimple_stocks.loc[i, 'EV/EBITDA'] = float(stock_stats.loc['Enterprise Value/EBITDA', 1])
    except:
        wealthsimple_stocks.loc[i, 'EV/EBITDA'] = np.NaN
    try:
        wealthsimple_stocks.loc[i, 'EV/Revenue'] = float(stock_stats.loc['Enterprise Value/Revenue', 1])
    except:
        wealthsimple_stocks.loc[i, 'EV/Revenue'] = np.NaN
    
    print(str(i) +' '+symbol+' '+str(count))
    time.sleep(1)
    count += 1
    if(count >= 90):
        time.sleep(300)
        count = 0
```
