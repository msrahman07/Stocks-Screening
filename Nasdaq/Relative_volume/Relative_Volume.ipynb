## Calculate change in volume

```python
#new incoming df
new_df = pd.read_csv('america_'+today+' ('+str(divider)+')'+'.csv')
new_df = new_df.drop(columns=['Description'])

#==================================================
change_tracker_df['Sum_of_price_change'] += ((new_df['Price'] - initial_df['Price'])/initial_df['Price'])*100
change_tracker_df['Sum_of_volume_change'] += ((new_df['Volume'] - initial_df['Volume'])/initial_df['Volume'])*100
change_tracker_df['Current_price_change'] = ((new_df['Price'] - initial_df['Price'])/initial_df['Price'])*100
change_tracker_df['Current_volume_change'] = ((new_df['Volume'] - initial_df['Volume'])/initial_df['Volume'])*100
change_tracker_df['Avg_price_change'] = change_tracker_df['Sum_of_price_change'] / divider
change_tracker_df['Avg_volume_change'] = change_tracker_df['Sum_of_volume_change'] / divider
# change_tracker_df['Volume_diff'] = change_tracker_df['Current_volume_change']-change_tracker_df['Avg_volume_change']
divider += 1
initial_df = new_df
#==================================================
rel_vol_df = pd.DataFrame()
rel_vol_df['Ticker'] = new_df['Ticker']
rel_vol_df['Price'] = new_df['Price']
rel_vol_df['Price_diff'] = change_tracker_df['Current_price_change'] - change_tracker_df['Avg_price_change']
rel_vol_df['Volume_diff'] = change_tracker_df['Current_volume_change'] - change_tracker_df['Avg_volume_change']
#==================================================

# average volume change
# percentage diff from current volume and average volume change
# change_tracker_df.sort_values(by='Volume_diff', ascending=False)[:50]
```

```python
rel_vol_df = rel_vol_df.sort_values(by='Volume_diff', ascending=False)[:50]
rel_vol_df = rel_vol_df.reset_index(drop=True)
rel_vol_df
```
